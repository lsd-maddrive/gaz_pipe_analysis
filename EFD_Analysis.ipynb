```python
import os
import pandas as pd
import struct as st
import xml.etree.ElementTree as ET
```

```python
# Before work put files 0.efd and DescrInfo.xml to folder

root = ET.parse('DescrInfo.xml').getroot()

num_sens = int(root.find('NumSens').text)
num_sens_idod = int(root.find('NumSensIDOD').text)
num_sens_f = int(root.find('NumSensF').text)

is_topo_correct = bool(int(root.find('Topograph').text))

print(num_sens, num_sens_idod, num_sens_f, is_topo_correct)
```

```python
FPATH = '0.efd'

file_sz = os.path.getsize(FPATH)
# print(file_sz)

pack_sz = file_sz//100000
print(pack_sz)

with open(FPATH, 'rb') as f:    
    
    for i in range(10):
        print('--------')
        sens_off = 0
        sens_sz = num_sens*2

        sens_idod_off = sens_off + sens_sz
        sens_idod_sz = num_sens_idod*2
        sens_f_off = sens_idod_off + sens_idod_sz
        sens_f_sz = num_sens_f*2
        info_off = sens_f_off + sens_f_sz
        info_sz = 73

        package_size = sens_sz + sens_idod_sz + sens_f_sz + info_sz
        package_data = f.read(package_size)

        sens_data = package_data[sens_off:sens_off+sens_sz]
        sens_idod_data = package_data[sens_idod_off:sens_idod_off+sens_idod_sz]
        sens_f_data = package_data[sens_f_off:sens_f_off+sens_f_sz]
        info_data = package_data[info_off:info_off+info_sz]

        # Info parsing
        dist_time = st.unpack('I5s', info_data[:9])
        package_data = st.unpack('9h', info_data[9:27])
        three_data = st.unpack('3b', info_data[27:30])
        # Ignore 29 bytes
        odom_data = st.unpack('3i', info_data[59:71])
        
        distance_mm = dist_time[0]
        time_data = dist_time[1]
        time_value = int.from_bytes(time_data, byteorder='little')
        print(f'Distance [mm]: {distance_mm}')
        print(f'Time [s]: {time_value/1e6}')
        
        print(f'Sensor number [on 12 h]: {package_data[0]}')
        print(f'Thickness [ADC]: {package_data[1]}')
        print(f'Speed [cm/s]: {package_data[2]}')
        print(f'Accel X: {package_data[3]}')
        print(f'Accel Y: {package_data[4]}')
        print(f'Accel Z: {package_data[5]}')
        print(f'Gyro X: {package_data[6]}')
        print(f'Gyro Y: {package_data[7]}')
        print(f'Gyro Z: {package_data[8]}')
        
        print(f'Temp [deg]: {three_data[0]}')
        print(f'Battery voltage [V]: {three_data[1]}')
        print(f'EFD version: {three_data[2]}')

        print(f'Odom [count]: {odom_data}')
    
```
